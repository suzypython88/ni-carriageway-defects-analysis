# Summary of Findings and Recommendations

* There is a large volume of road and path defects reported to the Northern Ireland Department for Infrastructure on an ongoing basis - in the 8 months between April and November 2025, 69,243 issues were raised.
* The majority of defects were bitmac and asphalt potholes on public roads (72%), largely occurring on Unclassified (55%) and Class C (17%), lesser maintained roads highlighting the need for monitoring, preventative measures and investment in higher quality road materials.
* Overall, 75% of issues reported were successfully repaired in this timeframe, with 77% of these were completed within the estimated timeframes given.
* Completion rates should be improved considering that amongst the 25% that remained unaddressed, communities have been left for up to 200 days with hazardous road conditions causing damage to private vehicles and making driving in these areas much more unsafe.
* Across the four regional divisions, the Southern (30.7%) and Western (31%) Divisions received the largest volumes of defect reports, especially for Unclassified and Class C roads.
* The Southern Division requires some investigation into staffing and efficiencies, as their overall completion rate was only 67%, with un fixed repairs remaining a hazard for on average 4 months (120 days).
  * Within this division, Armagh City, Banbridge and Craigavon East and West both addressed less than 50% of reported defects, and projects running over on average 23 days compared to estimates (East section office).
* The Western Division similarly requires some oversight due to large volumes and apparent capacity issues. Mid Ulster North had average completion times of a month, double that of any other office in this region, typically running over by 2 weeks compared to estimates.

Overall, there are clear performance issues given the volume of incoming road issues and the level of response capable by the section offices in each division. Funding is required to see that all major issues can be addressed in a timely manner, to avoid casualties and a massive private motor damage fund that remains necessary.

# Data Wrangling: Calculate Time Since Defect Recorded

Calculate time since an issue was recorded (if it does not have a completion date) or how much time it took to complete (if there is a completion date).

```python
# Calculate Time_Since_Recorded (in Days)

# Find most recent date in dataset
max_date = df4['Recorded_Date_Time'].max()
max_date

# If there is a completion date, calculate the difference between the recorded date and the completion date
# If there is no completion date, calculate the difference between the recorded date and the most recent date in the dataframe
df4['Time_Since_Recorded'] = np.where(
    df4['Completion_Date_Time'].notnull() == True, (df4['Completion_Date_Time'] - df4['Recorded_Date_Time']),
    (max_date - df4['Recorded_Date_Time']))

df4.head(20)
```

# Identify which Repairs have Exceeded Response Estimate Times
1. Calculate the Maximum Time of each Response Category

```python
# What are the reponse time bands?
df4['RESPONSE_TIME_NAME'].value_counts()
```

I will calculate the maximum/upper limit of the response time to compare with actual response times to see which repairs and how many have run over.

* R1 - 1 Calendar Day = 1 Day
* R2 - 5 Working Days = 7 Days (adding 2 for weekend days)
* R2 - Weekend = 7 Days (assuming this means 'by the next weekend')
* R3 - 4 Weeks = 28 Days
* 6 - Weeks = 42 Days

The following will be set to null as no upper limit it given:
* R4 - next available programme
* R0 - As Soon as Practical
* From Matrix


```python
# Set the column 'Max_Response' to hold the Upper end of the repair response time estimate

conditions = [
    df4['RESPONSE_TIME_NAME'] == 'R1 - 1 Calendar Day',
    df4['RESPONSE_TIME_NAME'] == 'R2 - 5 Working days',
    df4['RESPONSE_TIME_NAME'] == 'R2 Weekend',
    df4['RESPONSE_TIME_NAME'] == 'R3 - 4 Weeks',
    df4['RESPONSE_TIME_NAME'] == '6 Weeks',
]

values = [1, 7, 7, 28, 42]

df4['Max_Response'] = np.select(conditions, values).astype(int)

df4.head()
df4.info() # Check the new column is an Integer

```

```python
# Check that the max times have been assigned correctly and the remaining categories are nulls

df4['Max_Response'].value_counts() # These correspond correctly to the above categories
```

```python
# Replace the '0's with 999 because this should represenent an absence of information, but would otherwise block classifying this column as an integer
df4['Max_Response'].replace(0, 999, inplace=True)


df4['Max_Response'].value_counts()
```

###2. Calculate the difference between estimated response times and actual response times

```python
# Convert Time_Since_Recorded to whole days
#df4['Time_Since_Recorded'] = td.total_days(df4['Time_Since_Recorded'])

# Time_Since_Recorded - Max_Response
df4['Actual_Response_Difference'] = (df4['Time_Since_Recorded'].dt.days - df4['Max_Response'])

df4.head()

```

```python
# Set response time difference to 999 for rows with Max_Response = 999 as this is not a valid response time
df4.loc[df4['Max_Response'] == 999, 'Actual_Response_Difference'] = 999

df4[df4['Max_Response'] == 999].head()
```

###NOTE: Any analysis on response time will focus only on defects that were given a numeric response time estimate - a filter will be applied first.

```python
# Create a new column to signal if a response time was under or over estimate
df4['Response_Estimate'] = np.where(
    df4['Actual_Response_Difference'] <= 0, 'Under', 'Over')

df4.head()
```

```python
pavement_analysis1 = pd.DataFrame(df6.groupby('Pavement_Type')['INSTRUCTION_REFERENCE'].count())
```

```python
# Set the global plot style theme
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6,6))
plt.ylim(0,70000)

# Plot barplot
pav1 = sns.barplot(x = 'Pavement_Type', y = 'INSTRUCTION_REFERENCE', data = pavement_analysis1)

# Set title & axis labels
pav1.set(title = 'Road and Path Defects by Pavement Type')
pav1.set(xlabel = 'Pavement Type')
pav1.set(ylabel = 'Number of Defects')

# Format y-axis with thousands separator
pav1.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

# Add percentages to the chart
# Calculate the total count for each container
totals = []
for container in pav1.containers:
    total = 0
    for bar in container:
        total += bar.get_height()
    totals.append(total)

# Add value labels as percentages inside the bar
for i, container in enumerate(pav1.containers):
    for j, bar in enumerate(container):
        pav1.annotate(str(round(bar.get_height()/totals[i]*100, 1)) + '%',
                    xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Show the plot
plt.show()
```

The vast majority of defects are occurring on public roads, with a number also on public paths.

# Summary of Findings and Recommendations

* There is a large volume of road and path defects reported to the Northern Ireland Department for Infrastructure on an ongoing basis - in the 8 months between April and November 2025, 69,243 issues were raised.
* The majority of defects were bitmac and asphalt potholes on public roads (72%), largely occurring on Unclassified (55%) and Class C (17%), lesser maintained roads highlighting the need for monitoring, preventative measures and investment in higher quality road materials.
* Overall, 75% of issues reported were successfully repaired in this timeframe, with 77% of these were completed within the estimated timeframes given.
* Completions rates should be improved considering that amongst the 25% that remained unaddressed, communities have been left for up to 200 days with hazardous road conditions causing damage to private vehicles and making driving in these areas much more unsafe.
* Across the four regional divisions, the Southern (30.7%) and Western (31%) Divisions received the largest volumes of defect reports, especially for Unclassified and Class C roads.
* The Southern Division requires some investigation into staffing and efficiencies, as their overall completion rate was only 67%, with un fixed repairs remaining a hazard for on average 4 months (120 days).
  * Within this division, Armagh City, Banbridge and Craigavon East and West both addressed less than 50% of reported defects, and projects running over on average 23 days compared to estimates (East section office).
* The Western Division similarly requires some oversight due to large volumes and apparent capacity issues. Mid Ulster North had average completion times of a month, double that of any other office in this region, typically running over by 2 weeks compare to estimates.

Overall, there are clear performance issues given the volume of incoming road issues and the level of response capable by the section offices in each division. Funding is required to see that all major issues can be addressed in a timely manner, to avoid casualties and a massive private motor damage fund that remains necessary.
